```python
data['price'].mean()
```

```python
print('el precio medio de las casas es {} '.format(data['price'].mean()))
```

# Cual es el precio medio de las viviendas con 2 baños

```python
data[data['bathrooms']== 2]['price'].mean()
```

### Buscar que es loc y iloc

```python
data.loc[data['bathrooms']==2,'price'].mean()
```

```python
print('El precio medio de las viviendas con 2 baños es {}'.format(round(data.loc[data['bathrooms']==2,'price'].mean(),2)))
```

# El precio minimo de viviendas con 3 cuartos

```python
data.loc[data['bathrooms']==3,'price'].min()
```

```python
print('El precio minimo de las viviendas con 3 baños es {}'.format(data.loc[data['bathrooms']==3,'price'].min(),2))
```
